K Means clustering is an unsupervised learning algorithm that will attempt to group similar clusters together in your data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs

# Make artifical dataset

data = make_blobs(n_samples=200,
n_features=2,
centers=4,
cluster_std=1.8,
random_state=101
)
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1], cmap='rainbow')
```

```python
data[1]
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4)

kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, sharey=True, figsize=(10,6))

ax1.set_title('K Means Output')
ax1.scatter(data[0][:,0], data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original data')
ax2.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```
